```python
autos['nr_of_pictures'].value_counts()
```

The 'nr_of_pictures' column has all its values to be 0 so we will have to drop this column. We will also drop the 'seller'and 'offer_type' columns because they each have mostly the same values.

```python
autos.drop(['nr_of_pictures', 'seller', 'offer_type'], axis=1)
```

```python
autos["price"] = autos["price"].str.replace('$', '').str.replace(',', '') .astype(int)
                           
autos['price'].head()
```

```python
autos['odometer'] = autos['odometer'].str.replace('km','').str.replace(',', '').astype(int)
autos.rename({'odometer': 'odometer_km'}, axis = 1, inplace=True)
autos['odometer_km'].head()
```

```python
autos['odometer_km'].value_counts()
```

From the count above, you can observe that cars with the highest number of milleage has the highest number of counts. This means that there are more cars with high milleage than cars with low milleage.

```python
print(autos['price'].unique().shape)
print(autos['price'].describe())
print(autos['price'].value_counts().head(20))
```

There are 1421 car listings with a price of 0 which is lie 2% 0f the total number of cars. we can remove these rows. We can also see that the highest price is one hendred milloin which seems very high so let's look at the highest and lowest prices further.

There are a number of listings with ridiculously high prices of around and above one million including 14. We will remove these rows since the other prices seem to increase normally upto 350000 before going up very high to prices that are less realistic.

```python
autos['price'].value_counts().sort_index(ascending=True).head(20)
```

There are a number of listings with prices below 30 including around 1500 for 0. We will have to remove rows with such prices but we will leave rows from 1 above considering that eBay is an auction site and biddings might start from 1.

```python
autos = autos[autos['price'].between(1,351000)]
autos['price'].describe()
```

The regitration year of the car indicates the age of the car but the minimum and maximum values from the column description is odd. It has min of 1000 which is years befor cars were invented and max value of 9999 which is years into the future.

```python
(~autos['registration_year'].between(1900,2016)).sum() / autos.shape[0]
```

I picked a highest value of 2016 and lowest value of 1900 for registration year to be my range for valid data and calculated the percentage of my total data that has invalid data. And from the result above, the invalid data is of a very small percentage, therefore my decision is to remove this invalid data from the rgistration year column.

```python
autos = autos[autos['registration_year'].between(1900,2016)]
autos['registration_year'].value_counts(normalize=True).head(10)
```

Most of the vehicles were first registered 20years ago

```python
brand_count = autos['brand'].value_counts(normalize = True)
```

Volkswagen is the most popular brand by listing and there are a lot of other brands with insignificant percentage of listings. So I have decided to aggregate by brands with listings more than five percent of the total listing.

```python
common_brands = brand_count[brand_count > 0.05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    only_brand = autos[autos['brand'] == brand]
    mean_price = only_brand['price'].mean()
    brand_mean_prices[brand] = int(mean_price)
brand_mean_prices
```

Of the top 6 car brands, audi, mercedes benz and bmw are the most expensive respectively while ford and opel are less expensive. Volkswagen is in the middle, this could also explain why it is the most popular car brand because it is not too expensive and not too cheap.

```python
brand_mean_mileage = {}

for brand in common_brands:
    only_brand = autos[autos['brand'] == brand]
    mean_mileage = only_brand['odometer_km'].mean()
    brand_mean_mileage[brand] = int(mean_mileage)
    
mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_price = pd.Series(brand_mean_prices).sort_values(ascending=False)
```

```python
brand_info = pd.DataFrame(mean_mileage,columns = ['mean_mileage'])
brand_info
```

```python
brand_info['mean_price'] = mean_price
brand_info
```

The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
